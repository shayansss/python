## Basics

- Sequences data are typically variable-sized and not *independent and identically distributed* (***i.i.d.***), such as
    - text data,
    - time series,
    - nested sequences (e.g., videos), etc.
- Types of time series
    - by the number of values per time step:
        - ***Univariate***: e.g., a series with ***dimensionality***=1 and the shape [batch size, time steps, 1].
        - ***Multivariate***: e.g., measurement of revenue, debt.
    - by their tasks:
        - ***forecasting***
        - ***postdicting*** (filling the missing values by local prediction)
- When a sequence model uses the previous time-steps to predict the current state, it has the *autoregressive* property:
    - Given the input sequence data $X = x_{1:N} = \left \{  x_1, x_2, ..., x_N\right \}$, the output sequence $Y = y_{1:M} = \left \{  y_1, y_2, ..., y_M\right \}$ can be generated by the *chain rule of probability* that defines the output conditional probabilities $p(\mathbf{y}|\mathbf{x})$ using $\mathbf{y}_{<t}$ (i.e., output sequence data before the time step $t$):
\begin{equation}
p(\mathbf{y}|\mathbf{x})=\prod_{i=1}^{T}p\left ( y_{t}|\mathbf{y}_{<t}, \mathbf{x} \right ),
\end{equation}
    - This **autoregressive model (AR)** is a sequential function as each output sequence generation requires generation of the previous outputs ($\mathbf{y}_{<t}$), making it slow comparing to parralized non-AR models; but non-AR models may have sub-optimal performance [19].
    - The AR forecast problems are solved in statistics by techinques such as the ***weighted moving average*** or ***autoregressive integrated moving average (ARIMA)***, which may require us to first remove or handle components of systematic (that can be modeled) and non-systematic (e.g., noises) patterns, e.g., by decomposion of them. The systematic components include
        - ***Level***: the average value in the series.
        - ***growth***: the average increase or decrease observed in a period,
        - ***trends***: e.g., a continuous increase in the number of clients monthly;
        - ***seasonality***: e.g., the numbers that may change in each season (which can be handled by ***differencing*** that considers them by theirs differences at different days);
        - ***cyclicity***: similar to *seasonality* but without a fixed frequency.
    - Patterns like trends and seasonality, make the series non-starationary (time-dependent) and therefore more complicated for most statistical methods. Thus, they are usually converted to stationary data by techniques like differencing, complicating their implimentation [15].
    - Alternatively, such treatments in data with machine learning might be useful but not required. Common techiniques include:
        - *Recurrent AR* by **recurrent neural network (RNN)**, which approximates a function AR models by a recursive function that while sequentially computes the outputs but typically determine what information should be used for the next step as its *memory*. Its major challenges are:
            - ***Unstable gradients***, which can be alleviated by e.g., **recurrent dropout** and **layer normalization**.
            - ***limited short-term memory***, which can be extended by **LSTM** and **GRU cells**.
        - *Deep AR* by **convolusional neural networks (CNN)**, which uses feed-forward neural networks (FFNN) with easier training due to the possible application of non-saturating activation function and batch normalization, that handle very long sequences by considering hierarchical structure of sequences [20].
        - **Transformers**, which avoids recurrence (to redcuce repititions) and convolusions (to reduce very deep architectures) for long sequences, allow very efficient and parrallel implementation, mostly by different attension mechanisms (i.e., learnable methods that dynamically favor different parts of input data) [21].

## Recurrent neurons and layers
### Intuition
- As it has a **memory cell** (i.e., a ***recurrent neuron*** or ***layer***), it is not a FFNN.
- At time step $t$, i.e., a ***frame***, it can take pattern of typically previous 10 steps (or 100 steps for more powerful models) using the state of the previous step $\mathbf{h}_{\left(t\right)}$ (stands for “*hidden*”). Figure below shows the cell and its *unrolled* network through time:

<img src="img/rnn1.jpg" width="300px">
<center>The cell’s hidden state.</center>

In basic (not complex) cells, $\mathbf{h}_{\left(t\right)}=\mathbf{y}_{\left(t\right)}$:
- The input connections has weight vectors $\mathbf{w}_x$ and $\mathbf{w}_y$ and the ReLU and tanh can be used.
- For a vectorized layer of recurrent neurons with a minibatch, we have:

$\mathbf{Y}_{\left(t\right)}=\emptyset\left(\left[\begin{matrix}\mathbf{X}_{\left(t\right)}&\mathbf{Y}_{\left(t-1\right)}\\\end{matrix}\right]\left[\begin{matrix}\mathbf{W}_x\\\mathbf{W}_y\\\end{matrix}\right]+\mathbf{b}\right)$
- where:
  - $\mathbf{Y}_{\left(t\right)}$ is of $m\times n_\text{neurons}$ dimension ($m$ and $n_\text{neurons}$ are # mini-batch’s instances and neurons, respectively).
  - $\mathbf{X}_{\left(t\right)}$ is of $m\times n_\text{inputs}$ dimension.
  - $\mathbf{W}_x$ is of $n_\text{inputs}\times n_\text{neurons}$ dimension.
  - $\mathbf{W}_y$ is of $n_\text{neurons}\times n_\text{neurons}$ dimension.
  - $\mathbf{b}$ is of $n_\text{neurons}$ dimension.
  - $\emptyset$ is the ReLU or tanh fuctions.

```python
# Dataset generation:

def generate_time_series(batch_size, n_steps):
    freq1, freq2, offsets1, offsets2 = np.random.rand(4, batch_size, 1)
    time = np.linspace(0, 1, n_steps)
    series = 0.5 * np.sin((time - offsets1) * (freq1 * 10 + 10))  #   wave 1
    series += 0.2 * np.sin((time - offsets2) * (freq2 * 20 + 20)) # + wave 2
    series += 0.1 * (np.random.rand(batch_size, n_steps) - 0.5)   # + noise
    return series[..., np.newaxis].astype(np.float32)
```

```python
np.random.seed(42)

n_steps = 50
series = generate_time_series(10000, n_steps + 1)
X_train, y_train = series[:7000, :n_steps], series[:7000, -1]
X_valid, y_valid = series[7000:9000, :n_steps], series[7000:9000, -1]
X_test, y_test = series[9000:, :n_steps], series[9000:, -1]
```

```python
# samples of the valid sets:

def plot_series(series, y=None, y_pred=None, x_label="$t$", y_label="$x(t)$"):
    plt.plot(series, ".-")
    if y is not None:
        plt.plot(n_steps, y, "bx", markersize=10)
    if y_pred is not None:
        plt.plot(n_steps, y_pred, "ro")
    plt.grid(True)
    if x_label:
        plt.xlabel(x_label, fontsize=16)
    if y_label:
        plt.ylabel(y_label, fontsize=16, rotation=0)
    plt.hlines(0, 0, 100, linewidth=1)
    plt.axis([0, n_steps + 1, -1, 1])

fig, axes = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(12, 4))
for col in range(3):
    plt.sca(axes[col])
    plot_series(X_valid[col, :, 0], y_valid[col, 0],
                y_label=("$x(t)$" if col==0 else None))
save_fig("time_series_plot")
plt.show()
```

```python
# Linear predictions:

np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[50, 1]),
    keras.layers.Dense(1)
])

model.compile(loss="mse", optimizer="adam")
history = model.fit(X_train, y_train, epochs=20,
                    validation_data=(X_valid, y_valid))
```

```python
model.evaluate(X_valid, y_valid)
```

```python
def plot_learning_curves(loss, val_loss):
    plt.plot(np.arange(len(loss)) + 0.5, loss, "b.-", label="Training loss")
    plt.plot(np.arange(len(val_loss)) + 1, val_loss, "r.-", label="Validation loss")
    plt.gca().xaxis.set_major_locator(mpl.ticker.MaxNLocator(integer=True))
    plt.axis([1, 20, 0, 0.05])
    plt.legend(fontsize=14)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.grid(True)

plot_learning_curves(history.history["loss"], history.history["val_loss"])
plt.show()
```

```python
def plot_multiple_forecasts(X, Y, Y_pred):
    n_steps = X.shape[1]
    ahead = Y.shape[1]
    plot_series(X[0, :, 0])
    plt.plot(np.arange(n_steps, n_steps + ahead), Y[0, :, 0], "ro-", label="Actual")
    plt.plot(np.arange(n_steps, n_steps + ahead), Y_pred[0, :, 0], "bx-", label="Forecast", markersize=10)
    plt.axis([0, n_steps + ahead, -1, 1])
    plt.legend(fontsize=14)

plot_multiple_forecasts(X_new, Y_new, Y_pred)
plt.show()
```

- predicts series of outputs from the beginning by a sequence to sequence RNN. This way, the model inputs and outputs may overlap a lot, which is not an issue as it is a ***casual*** model (i.e., it remembers only past not future). For this
    - we can use `TimeDistributed()`, a wrapper to apply a layer to every temporal slice of the input like a convolutional layer applied to the time axis usually axis 1.
    - For evaluation, we use only the values at the last time steps. So, a custom metric should be defined.
    - Use some error bars along with predictions, e.g., by firstly adding MC dropout on each layer. Then, during predictions, compute different statistics at each time step.
- The above models can be combined to generate mildly long sequences, e.g., texts or music. 

```python
# An RNN that predicts the next 10 steps at each time step

np.random.seed(42)

n_steps = 50
series = generate_time_series(10000, n_steps + 10)
X_train = series[:7000, :n_steps]
X_valid = series[7000:9000, :n_steps]
X_test = series[9000:, :n_steps]
Y = np.empty((10000, n_steps, 10))
for step_ahead in range(1, 10 + 1):
    Y[..., step_ahead - 1] = series[..., step_ahead:step_ahead + n_steps, 0]
Y_train = Y[:7000]
Y_valid = Y[7000:9000]
Y_test = Y[9000:]

X_train.shape, Y_train.shape
```

```python
np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.SimpleRNN(20, return_sequences=True, input_shape=[None, 1]),
    keras.layers.SimpleRNN(20, return_sequences=True),
    keras.layers.TimeDistributed(keras.layers.Dense(10))
])

# `TimeDistributed(Dense(n))` is equivalent to Dense(n), since it only applies to the last axis.
# But we use it to emphasise that the models at each time step are separated to each other.

def last_time_step_mse(Y_true, Y_pred):
    return keras.metrics.mean_squared_error(Y_true[:, -1], Y_pred[:, -1])

model.compile(loss="mse", optimizer=keras.optimizers.Adam(lr=0.01), metrics=[last_time_step_mse])
history = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

## Unstable gradients

The tricks used for Deep NN can also be used here, except that:
- Nonsaturating activation functions may not work [1], as the same weights are used for all time steps which can lead to explosion of outputs or gradient. You can also check the lr.
- the benefits of the BN layer can be limited (as it does not change the statistics in each time step).

```python
# Deep RNN with Batch Norm

np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.SimpleRNN(20, return_sequences=True, input_shape=[None, 1]),
    keras.layers.BatchNormalization(),
    keras.layers.SimpleRNN(20, return_sequences=True),
    keras.layers.BatchNormalization(),
    keras.layers.TimeDistributed(keras.layers.Dense(10))
])

model.compile(loss="mse", optimizer="adam", metrics=[last_time_step_mse])
history = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

- Instead of BN, Layer Normalization [2] may be more useful that normalizes it across the feature dimensions (rather than the batch). This can be implemented by a custom memory cell on the fly, and it does not need the weighted average.
- We use  `LayerNormalization` in a custom class, which is very similar to the regular RNN, except for its `call()` method that also updates the current inputs and states (in a simple RNN these two are equal).
- We may also use either the `recurrent_dropout` (for hidden states) and dropout (for inputs).

```python
# Deep RNNs with Layer Norm

from tensorflow.keras.layers import LayerNormalization

class LNSimpleRNNCell(keras.layers.Layer):
    def __init__(self, units, activation="tanh", **kwargs):
        super().__init__(**kwargs)
        self.state_size = units
        self.output_size = units
        self.simple_rnn_cell = keras.layers.SimpleRNNCell(units,
                                                          activation=None)
        self.layer_norm = LayerNormalization()
        self.activation = keras.activations.get(activation)
    def get_initial_state(self, inputs=None, batch_size=None, dtype=None):
        if inputs is not None:
            batch_size = tf.shape(inputs)[0]
            dtype = inputs.dtype
        return [tf.zeros([batch_size, self.state_size], dtype=dtype)]
    def call(self, inputs, states):
        outputs, new_states = self.simple_rnn_cell(inputs, states)
        norm_outputs = self.activation(self.layer_norm(outputs))
        return norm_outputs, [norm_outputs]
```

```python
np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.RNN(LNSimpleRNNCell(20), return_sequences=True,
                     input_shape=[None, 1]),
    keras.layers.RNN(LNSimpleRNNCell(20), return_sequences=True),
    keras.layers.TimeDistributed(keras.layers.Dense(10))
])

model.compile(loss="mse", optimizer="adam", metrics=[last_time_step_mse])
history = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

## Short-term memory problem

### LSTM cells
- The **Long Short-Term Memory (LSTM) cell** may be better and even faster to converge.
- To implement it,
    - use `LSTMCell()` in an `RNN()`, or
    - simply use `LSTM()`, which is more optimized for GPU implementation.
- As shown in figure below, here, the output state is split into short-term state $\mathbf{h}_{\left(t\right)}$ and long-term state $\mathbf{c}_{\left(t\right)}$ ($\mathbf{c}$ means cell).

<img src="img/lstm.jpg" width="400px">
<center>An LSTM cell.</center>

- FC blocks denotes fully connected layers: the major block is $\mathbf{g}_{\left(t\right)}$ and the others are gate controllers.
- Below are the LSTM computations:

\begin{equation}
\mathbf{i}_{\left(t\right)}=\sigma\left({\mathbf{W}_{xi}}^T\mathbf{x}_{\left(t\right)}+{\mathbf{W}_{hi}}^T\mathbf{h}_{\left(t-1\right)}+\mathbf{b}_i\right)\\
\mathbf{f}_{\left(t\right)}=\sigma\left({\mathbf{W}_{xf}}^T\mathbf{x}_{\left(t\right)}+{\mathbf{W}_{hf}}^T\mathbf{h}_{\left(t-1\right)}+\mathbf{b}_f\right)\\
\mathbf{o}_{\left(t\right)}=\sigma\left({\mathbf{W}_{xo}}^T\mathbf{x}_{\left(t\right)}+{\mathbf{W}_{ho}}^T\mathbf{h}_{\left(t-1\right)}+\mathbf{b}_o\right)\\
\mathbf{g}_{\left(t\right)}=\text{tanh}\left({\mathbf{W}_{xg}}^T\mathbf{x}_{\left(t\right)}+{\mathbf{W}_{hg}}^T\mathbf{h}_{\left(t-1\right)}+\mathbf{b}_g\right)\\
\mathbf{c}_{\left(t\right)}=\mathbf{f}_{\left(t\right)}\otimes\mathbf{c}_{\left(t-1\right)}+\mathbf{i}_{\left(t\right)}\otimes\mathbf{g}_{\left(t\right)}\\
\mathbf{y}_{\left(t\right)}=\mathbf{h}_{\left(t\right)}=\mathbf{o}_{\left(t\right)}\text{tanh}\left(\mathbf{c}_{\left(t\right)}\right)
\end{equation}
- where, $\otimes$ is the element-wise multiplication operator, and $\sigma$ is the sigmoid function.
- The tanh function keeps outputs between -1 to +1, while the sigmoid function keeps values between 0 to 1 for gates. Thus, in TF, $\mathbf{b}_f$ is initialized with a vector of 1 to forget everything at first.

```python
np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.LSTM(20, return_sequences=True, input_shape=[None, 1]),
    keras.layers.LSTM(20, return_sequences=True),
    keras.layers.TimeDistributed(keras.layers.Dense(10))
])

model.compile(loss="mse", optimizer="adam", metrics=[last_time_step_mse])
history = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

### Peephole connections
- It is a variant of LSTM with peephole connections, which
    - feed $\mathbf{c}_{\left(t-1\right)}$ into the controllers, i.e., the FC blocks, of the forget and input gates and also
    - feed $\mathbf{c}_{\left(t\right)}$ into the controller of the output gate.
- This way, it may or may not work better.
- For implementation, use pass `experimental.PeepholeLSTMCell` to the constructor of the `RNN` layer.

### GRUs
**Gated Recurrent Unit (GRU)** is another variant of LSTM that may performs just as well with these differences:
- State vector are merged into one vector.
- The gait controller $z$ controls both input and forget gaits by **update gate**.
- There is not an output gait, but the new gait $\mathbf{r}_{\left(t\right)}$ controls which part of the previous state can be transferred to the main layer $\mathbf{g}_{\left(t\right)}$.

<img src="img/gru.jpg" width="400px">
<center>An GRU cell.</center>

- Here, the equations are

\begin{equation}
\mathbf{z}_{\left(t\right)}=\sigma\left({\mathbf{W}_{xz}}^T\mathbf{x}_{\left(t\right)}+{\mathbf{W}_{hz}}^T\mathbf{h}_{\left(t-1\right)}+\mathbf{b}_z\right)\\
\mathbf{r}_{\left(t\right)}=\sigma\left({\mathbf{W}_{xr}}^T\mathbf{x}_{\left(t\right)}+{\mathbf{W}_{hr}}^T\mathbf{h}_{\left(t-1\right)}+\mathbf{b}_r\right)\\
\mathbf{g}_{\left(t\right)}=\text{tanh}\left({\mathbf{W}_{xg}}^T\mathbf{x}_{\left(t\right)}+{\mathbf{W}_{hg}}^T\left(\mathbf{r}_{\left(t\right)}\otimes\mathbf{h}_{\left(t-1\right)}\right)+\mathbf{b}_g\right)\\
\mathbf{h}_{\left(t\right)}=\mathbf{z}_{\left(t\right)}\otimes\mathbf{h}_{\left(t-1\right)}+\left(\mathbf{1}-\mathbf{z}_{\left(t\right)}\right)\otimes\mathbf{g}_{\left(t\right)}
\end{equation}

- To implement it, use `GRU`, which is based on `layers.GRUCell`.

```python
np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.GRU(20, return_sequences=True, input_shape=[None, 1]),
    keras.layers.GRU(20, return_sequences=True),
    keras.layers.TimeDistributed(keras.layers.Dense(10))
])

model.compile(loss="mse", optimizer="adam", metrics=[last_time_step_mse])
history = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

#### WaveNet

- It uses the ***causal padding*** (for left padding with zeros to avoid smaller and smaller sequences), as shown below.

<center>Comparing same/valid padding (left) with causal padding (right) [4].</center>

- 1d convolutions with causal padding and dilation are often referred to as ***temporal convolution*** [4].

```python
# simplified implementation

np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential()
model.add(keras.layers.InputLayer(input_shape=[None, 1]))
for rate in (1, 2, 4, 8) * 2:
    model.add(keras.layers.Conv1D(filters=20, kernel_size=2, padding="causal",
                                  activation="relu", dilation_rate=rate))
model.add(keras.layers.Conv1D(filters=10, kernel_size=1))
model.compile(loss="mse", optimizer="adam", metrics=[last_time_step_mse])
history = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

- The original WaveNet uses ***resedual connections*** [5] and ***skip connections*** to speed up training (figure below).

<img src="img/original_wavenet.jpg" width="500px">
<center>Original WaveNet [3].</center>

- Also it uses the ***gated activation unit*** (originally defined in the gated PixelCNN [6]), which they found outperfoms the ReLU in the case of modeling audio signals:

\begin{equation}
\mathbf{z} =  \text{tanh} \left( {W}_{f,k} \ast \mathbf{x} \right) \otimes \sigma \left( {W}_{g,k} \ast \mathbf{x} \right),
\end{equation}

- where, $\ast$ denotes the convolution operator, k denotes the layer index, f and g are respectively the filter and gate, and W is a learnable convolution filter.

```python
class GatedActivationUnit(keras.layers.Layer):
    def __init__(self, activation="tanh", **kwargs):
        super().__init__(**kwargs)
        self.activation = keras.activations.get(activation)
    def call(self, inputs):
        n_filters = inputs.shape[-1] // 2
        linear_output = self.activation(inputs[..., :n_filters])
        gate = keras.activations.sigmoid(inputs[..., n_filters:])
        return self.activation(linear_output) * gate
```

```python
def wavenet_residual_block(inputs, n_filters, dilation_rate):
    z = keras.layers.Conv1D(2 * n_filters, kernel_size=2, padding="causal",
                            dilation_rate=dilation_rate)(inputs)
    z = GatedActivationUnit()(z)
    z = keras.layers.Conv1D(n_filters, kernel_size=1)(z)
    return keras.layers.Add()([z, inputs]), z
```

```python
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)

n_layers_per_block = 3 # 10 in the paper
n_blocks = 1 # 3 in the paper
n_filters = 32 # 128 in the paper
n_outputs = 10 # 256 in the paper

inputs = keras.layers.Input(shape=[None, 1])
z = keras.layers.Conv1D(n_filters, kernel_size=2, padding="causal")(inputs)
skip_to_last = []
for dilation_rate in [2**i for i in range(n_layers_per_block)] * n_blocks:
    z, skip = wavenet_residual_block(z, n_filters, dilation_rate)
    skip_to_last.append(skip)
z = keras.activations.relu(keras.layers.Add()(skip_to_last))
z = keras.layers.Conv1D(n_filters, kernel_size=1, activation="relu")(z)
Y_proba = keras.layers.Conv1D(n_outputs, kernel_size=1, activation="softmax")(z)

model = keras.models.Model(inputs=[inputs], outputs=[Y_proba])
```

```python
model.compile(loss="mse", optimizer="adam", metrics=[last_time_step_mse])
history = model.fit(X_train, Y_train, epochs=2,
                    validation_data=(X_valid, Y_valid))
```

## Customization

### Custom RNN cell

- E.g., when nested structures like `[batch, timestep, {"video": [height, width, channel], "audio": [frequency]}]` are implemented.

```python
# E.g., using random Numpy data (from https://www.tensorflow.org/guide/keras/rnn)

unit_1 = 10
unit_2 = 20
unit_3 = 30

i1 = 32
i2 = 64
i3 = 32
batch_size = 64
num_batches = 10
timestep = 50

input_1_data = np.random.random((batch_size * num_batches, timestep, i1))
input_2_data = np.random.random((batch_size * num_batches, timestep, i2, i3))
target_1_data = np.random.random((batch_size * num_batches, unit_1))
target_2_data = np.random.random((batch_size * num_batches, unit_2, unit_3))
```

```python
class NestedCell(keras.layers.Layer):
    def __init__(self, unit_1, unit_2, unit_3, **kwargs):
        self.unit_1 = unit_1
        self.unit_2 = unit_2
        self.unit_3 = unit_3
        self.state_size = [tf.TensorShape([unit_1]), tf.TensorShape([unit_2, unit_3])]
        self.output_size = [tf.TensorShape([unit_1]), tf.TensorShape([unit_2, unit_3])]
        super(NestedCell, self).__init__(**kwargs)

    def build(self, input_shapes):
        # expect input_shape to contain 2 items, [(batch, i1), (batch, i2, i3)]
        i1 = input_shapes[0][1]
        i2 = input_shapes[1][1]
        i3 = input_shapes[1][2]

        self.kernel_1 = self.add_weight(
            shape=(i1, self.unit_1), initializer="uniform", name="kernel_1"
        )
        self.kernel_2_3 = self.add_weight(
            shape=(i2, i3, self.unit_2, self.unit_3),
            initializer="uniform",
            name="kernel_2_3",
        )

    def call(self, inputs, states):
        # inputs should be in [(batch, input_1), (batch, input_2, input_3)]
        # state should be in shape [(batch, unit_1), (batch, unit_2, unit_3)]
        input_1, input_2 = tf.nest.flatten(inputs)
        s1, s2 = states

        output_1 = tf.matmul(input_1, self.kernel_1)
        output_2_3 = tf.einsum("bij,ijkl->bkl", input_2, self.kernel_2_3)
        state_1 = s1 + output_1
        state_2_3 = s2 + output_2_3

        output = (output_1, output_2_3)
        new_states = (state_1, state_2_3)

        return output, new_states

    def get_config(self):
        return {"unit_1": self.unit_1, "unit_2": unit_2, "unit_3": self.unit_3}
```

```python
cell = NestedCell(unit_1, unit_2, unit_3)
rnn = keras.layers.RNN(cell)

input_1 = keras.Input((None, i1))
input_2 = keras.Input((None, i2, i3))

outputs = rnn((input_1, input_2))

model = keras.models.Model([input_1, input_2], outputs)

model.compile(optimizer="adam", loss="mse", metrics=["accuracy"])
```

```python
input_data = [input_1_data, input_2_data]
target_data = [target_1_data, target_2_data]

model.fit(input_data, target_data, batch_size=batch_size)
```

### Custom RNN layer

```python
# for demonstration (from the 3rd ed. of Geron's book)

class MyRNN(tf.keras.layers.Layer):
    def __init__(self, cell, return_sequences=False, **kwargs):
        super().__init__(**kwargs)
        self.cell = cell
        self.return_sequences = return_sequences

    def get_initial_state(self, inputs):
        try:
            return self.cell.get_initial_state(inputs)
        except AttributeError:
            # fallback to zeros if self.cell has no get_initial_state() method
            batch_size = tf.shape(inputs)[0]
            return [tf.zeros([batch_size, self.cell.state_size],
                             dtype=inputs.dtype)]

    @tf.function
    def call(self, inputs):
        states = self.get_initial_state(inputs)
        shape = tf.shape(inputs)
        batch_size = shape[0]
        n_steps = shape[1]
        # `@tf.function` requires creating 'outputs' before the `for` loop.
        # When the function is converted to a graph, this will be pruned.
        # For a similar reason, 'sequences' is created befor the `if` statement.
        sequences = tf.TensorArray(
            inputs.dtype, size=(n_steps if self.return_sequences else 0))
        outputs = tf.zeros(shape=[batch_size, self.cell.output_size],
                           dtype=inputs.dtype)
        
        for step in tf.range(n_steps):
            outputs, states = self.cell(inputs[:, step], states)
            if self.return_sequences:
                sequences = sequences.write(step, outputs)

        if self.return_sequences:
            # returning the outputs of the shape [time steps, batch size, dims].
            return tf.transpose(sequences.stack(), [1, 0, 2])
        else:
            return outputs
```

## Advanced preprocessing considerations


### Avoiding  bias in the test set
- Sometimes, e.g., in **natural language processing (NLP)**, the training, validation, and test sets should not be overlapped, and even, it is a good idea to leave a gap between them, e.g., to avoid paragraph overlapping.
- On the other hand, for time series we may split in other dimensions rather than times, e.g., across different companies. But it is highly possible that they (in this case, companies) are correlated, making the test sets very biased. Therefore, generally splitting over time is more preferred, e.g., from 2012 to 2014 and from 2014 to 2016.

### Considering stationary and nonstationary data
- In *stationary* time series, the mean, variance, and ***autocorrelations*** (i.e., correlations between its separated intervals inside that time series) do not change over time, at least in the wide sense. E.g., chemical reactions (as their chemical laws do not change over time).
- To ensure that the time series is sufficiently stationary, plot the models’ errors on the validation set across time which should not change significantly.
- Time-series with cyclical patterns or particular trends are *nonstationary*, e.g., financial markets. For them, we may work on shorter time spans.

### Chopping sequences
- We can use `window()` to chop the dataset into smaller datasets (e.g., substrings) to avoid very large sequences.
- We may use `shift`, `drop_remainder`, `stride` arguments. The latter defines the stride between elements of each window.
- This result in a nested dataset, in which each window dataset can be operated seperately but should be converted to flat dataset (as training on nested datasets is not possible); thus, `flat_map()` can be used.

```python
# e.g., for demonstration

np.random.seed(42)
tf.random.set_seed(42)

# sequence 0 to 14 into windows of length 5, each shifted by 2:
n_steps = 5
dataset = tf.data.Dataset.from_tensor_slices(tf.range(10))
dataset = dataset.window(n_steps, shift=2, drop_remainder=True)

# here batch flat_map() makes each window dataset a batch.
dataset = dataset.flat_map(lambda window: window.batch(5))

for idx, window in enumerate(dataset):
    print('\n'+'Window ds '+str(idx)+':')
    for element in window:
        print(element)
```

### Stateless RNN
The ***stateful RNNs*** reserves the state for the next batch, enabling to learn long-term patterns. For the simplest implantation:

#### Making
- Using ***making***, the padding tokens are ignored by all downstream layers (so, the network does not need to learn it by itself). The mask tensor can be propagated to next layers, until the time dimension exists. If the mask reaches to the loss function, its loss is 0.
- For implementation either
    - Use the layers that can support it (check `supports_masking` attribute), and implement them by `mask_zero=True`, but note that they may not work with the Keras GPU optimization.
    - Use custom layers by adding a `mask` argument to its `call()`, and adding `self.supports_masking = True` in the constructor.
    - Use `keras.layers.Masking` if the layer does not begin with an Embedding layer: it mask out time steps where the last dimension is full of specifice values, e.g., zeros (equivalent to the implementation by`K.any(K.not_equal(inputs, 0), axis=-1)`).
    - It may not handle complex models, e.g., a combination of conv1D and RNN. For them explicitly compute the mask, and pass it the layer, using Functional or Subclassing API.

```python
K = keras.backend
embed_size = 128
inputs = keras.layers.Input(shape=[None])

# not_equal(inputs, 0)) return a (Boolean) mask tensor
mask = keras.layers.Lambda(lambda inputs: K.not_equal(inputs, 0))(inputs)

z = keras.layers.Embedding(vocab_size + num_oov_buckets, embed_size)(inputs)
z = keras.layers.GRU(128, return_sequences=True)(z, mask=mask)
z = keras.layers.GRU(128)(z, mask=mask)

# here the masks are not propogated, as no sequence is returned.
outputs = keras.layers.Dense(1, activation="sigmoid")(z)
model = keras.models.Model(inputs=[inputs], outputs=[outputs])
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
history = model.fit(train_set, epochs=5)
```

## An Encoder–Decoder Network for Neural Machine Translation

We start with a simple neural machine translation [7], as shown below:

<img src="img/translation.jpg" width="400px">
<center>A simple NLP method.</center>

- The encoder converts the varible-lenght input sequents to fixed-size ***context*** vector, i.e., the last encoder's hidden state containing the input's *contextual representation*.
- During training, the ground-trought target translations are fed (with one step delay) to the decoder (i.e., the ***Teacher Forcing*** algorithm [12]).
- At inference time, we use the predicted translations (as the true target translations are not available). This training-prediction discrepancy may cause ***Exposure Bias*** [13], i.e., when the training data distribution is different from the corresponding data distribution after traininig.
- The very first word is a *start-of-sequence* (**SOS**) token, and the decoder ends the sentence with an *end-of-sequence* (**EOS**) token.
- English sentences are reversed as verbs are more important.
- The decoder outputs a score, which turn into a probability by a softmax layer.

For implementation use the **TF Addons** project that has many *sequence-to-sequence* tools for production-ready encoder–decoders:
- Use the TF Addons project that has many sequence-to-sequence tools for production-ready encoder–decoders. Here, we may use `TrainingSampler()` that provides the decoder with the pretended  previous output (based on the predicted, or target values, as stated above). However, It is also suggested to start training with the target of the previous time step and gradually shift to using the actual values. For this, you may use `ScheduledEmbeddingTrainingSampler` that  randomly selects between the target or the predicted output, with a probability that can be changed gradually.
- As the sentences have varying lengths, we should group them into buckets of similar lengths, e.g., one bucket for 1- to 5-word sentences, another bucket for 6- to 10-word sentences, etc. Smaller sentences are simply padded to fill each bocket. Check `tf.data.experimental.bucket_by_sequence_length()`.
- Mask out (to ignore) any output pass the EOS token.
- As the output is too large, use `tf.nn.sampled_softmax_loss()` [37] that approximates the loss based on merely the logits outputs for the correct word and for some random incorrect words. However, at inference time (as targets are not available), use the regular softmax function.

## Attention mechanisms

- Attention mechanisms (firstly introduced by Graves et al. [25]), which can be with or without RNNs, can address the issue RNN short memory by focusing on the right word (instead of the regular RNN's encoders, that emphasises more on the input's hidden state).
- They are also helps to improve *explanaibility* by e.g., **visual attention** as used in ***image captions generation*** [22] (shown below).
<img src="img/attention_vis.jpg" width="400px">
<center>E.g., for the visual attention generating the caption “A woman is throwing a frisbee in a park”: an input image (left) and the model’s focus when it was close to the it was about to output the word “frisbee”.</center>
- They can be categorized to **inter-attention** and **intra-attention** mechanisms depending whether they consider the information of both inputs and outputs or just one of them.


### Attention mechanisms with RNNs or CNNs
As show in figure below:
<img src="img/attention.jpg" width="400px">
<center>An attention mechanism with RNNs.</center>

- In addition to transferring the final encoder’s hidden states (not shown here), the outputs are also sent to the decoder. A weighted sum of these outputs are computed using the weights $\alpha_{\left(t,i\right)}$ corresponding to the i$^\text{th}$ encoder output at the t$^\text{th}$ decoder time step.
- While the rest of the model works like other RNNs, weights are computed by another small NN, i.e., an ***alignment model*** (or an ***attention layer***) that is trained jointly with the encoder-decoder model.
- This is the **Bahdanau** attention mechanism [9]:
    - This alignment model starts with a time-distributed layer with a single neuron, receiving the concatenation of the previous decoder’s hidden layer and encoder’s outputs. Then, this layer outputs a ***score*** (or *energy*), i.e., $e$, for each, which measures how well each e is aligned with the decoder’s previous state. With the softmax, the weights sum to 1.
    - A simplification is to use $\mathbf{h}_{\left(t\right)}$ instead of $\mathbf{h}_{\left(t-1\right)}$.
    - This attention is also called ***concat*** [10] or ***additive*** [21] attention.
    - It is considered as the a ***global attention***  model, since it considers all inputs, 
    - It is ***soft attention*** as considers all the input hidden states by summation (in contrary to *hard attention* that strictly works based one hidden state).
- An alternative is the **Luong** attension [10]:
    - it measures the similarity by the *dot product* (thus, also callled ***multiplicative*** attention) to benefit from the hardware acceleration.
    - it has two implemenation, namely ***dot product*** and ***general dot product*** (for which, the outputs go through a linear transformation prior to the dot product).
    - The original implementation is a a ***local attention***  model, as it attends around only ***alignment position*** (instead of the whole sequence), to accelate model. The aligment position for $t^{th}$ decoder hidden state $p_t$ can be detemined by
        - ***monotonic alignment*** that assumes input and output sequences are relatively monotonically aligned, thus $p_t = t$;
        - ***predictive alignment*** that uses Gaussian distribution to favor alignment weights around the alignment position, which is then defined by a learable method: $L\cdot \text{sigmoid}\left ( v_{p}^{\text{T}} \text{tanh}\left ( W_p h_t \right )\right )$, where $L$ is the sequence length, and $W_p$ and $v_{p}$ are learnable parameters. 
- These attenstion mechanisms are compared bolow:

\begin{equation}
{\widetilde{\mathbf{h}}}_{\left(t\right)}=\sum_{i}{\alpha_{\left(t,i\right)}\mathbf{y}_{\left(i\right)}}
\quad \text{with} \quad
\alpha_{\left(t,i\right)}=\frac{exp\left(e_{\left(t,i\right)}\right)}{\sum_{i} e x p\left(e_{\left(t,i^\prime\right)}\right)}
\quad \text{and} \quad
e_{\left(t,i\right)}=\left\{\begin{matrix}{\mathbf{h}_{\left(t\right)}}^T\mathbf{y}_{\left(i\right)}&\text{dot}\\{\mathbf{h}_{\left(t\right)}}^T\mathbf{W}\mathbf{y}_{\left(i\right)}&\text{general}\\\mathbf{v}^T\mathbf{tanh}\left(\mathbf{W}\left[\mathbf{h}_{\left(t\right)};\mathbf{y}_{\left(i\right)}\right]\right)&\text{concat}\\\end{matrix}\right.
\end{equation}

- Here, $\mathbf{v}$ is a rescaling parameter vector.
- Generally, the dot products variants may work better and can be implemented by TF Addons.

- By the concept of **query**, **key**, **value**, the attention mechanism can be seen as a function outputing with weigthed sum of the *values*, where each weight is computed by a ***compatibility function*** of the query with the assosiated key [21].
- For illustration, in the above *encoder-docoder attention*, the queries are from each decoder's layer, which attents to all the key/values (which are usually the same) from the encoder [21].
- Acoordingly, the `Attention` layer implements the dot-product attention by this concept (that might be used as the base class of other attension layers):
    - The query, key, and value tensors are inputs to this layer with shapes `[batch_size, Tq or Tv or Tv, dimension]` (considering that values and keys are usually the same);
    - it calculate ***scores*** with shape `[batch_size, Tq, Tv]` by `scores = tf.matmul(query, key, transpose_b=True)`.
    - it then calculates a ***distribution*** with the same shape of scores by `distribution = tf.nn.softmax(scores)`.
    - it finally creates a linear combination of values of the shape `[batch_size, Tq, dim]` by `return tf.matmul(distribution, value)`.

### Transformers
- Google researchers suggested that “Attention Is All You Need.” [21], as their proposed method (figure below), i.e., transformers, can learn sequentional data mostly by attenstion mechanisms and related components without RNNs, which permits easier parallelization without the need for extremely deep CNN-based implementation for large sequences.
- The vanialla transformer is show below, which, among its components, it contains N sequential encoders and decoders.

<img src="img/transformer.jpg" width="300px">
<center>A transformer [21].</center>

- Since transformers do not have recurrence and convolusion operators, they use **positional encoding (PE)** [29], e.g., by the concept of Fourier series [28], to add positional information into fed data:
    - The vannilla transformer uses the following PE method of sequences:
    \begin{equation}
    {PE}_{\left(pos, 2i\right)}= \text{sin}(pos/10000^{2i/d_{\text{model}}}) \\
    {PE}_{\left(pos, 2i+1\right)}= \text{cos}(pos/10000^{2i/d_{\text{model}}})
    \end{equation}
    - Here,
        - $0\leq pos<L$ is the position of fed object in the sequence of length $L$,
        - Position function for mapping a position  in the input sequence to index  of the positional matrix,
        - the arbitrary number 10000 controls the frequency and wavelenght of of the trigonometric functions,
        - $0\leq i<d/2$ is an indice of used to for mapping the functions outputs to the components of the output embeddings with the dimension of $d$ (as illustrated below).
    <img src="img/PE3.png" width="500px">
    <center>E.g., of a positional encoding matrix for an input sentence [24].</center>
    - Mathematically speaking,
        - each dimension then corresponds to a sinusoid with different wavelengths, and
        - it may allow the optimizer to easily learn to attend when considering relative positions, as it can be proved that ${PE}_{\left(pos\right)}$, is linearly corrolated to ${PE}_{\left(pos+k\right)}$, for any arbitraty offset $k$.

```python
def positional_encoding(length, depth):
    depth = depth/2
    
    positions = np.arange(length)[:, np.newaxis]     # (seq, 1)
    depths = np.arange(depth)[np.newaxis, :]/depth   # (1, depth)
    
    angle_rates = 1 / (10000**depths)         # (1, depth)
    angle_rads = positions * angle_rates      # (pos, depth)
    
    pos_encoding = np.concatenate(
        [np.sin(angle_rads), np.cos(angle_rads)],
        axis=-1)
    
    return tf.cast(pos_encoding, dtype=tf.float32)

class PositionalEmbedding(tf.keras.layers.Layer):
    def __init__(self, vocab_size, d_model):
        super().__init__()
        self.d_model = d_model
        self.embedding = tf.keras.layers.Embedding(vocab_size, d_model, mask_zero=True) 
        self.pos_encoding = positional_encoding(length=2048, depth=d_model)
    
    def compute_mask(self, *args, **kwargs):
        return self.embedding.compute_mask(*args, **kwargs)
    
    def call(self, x):
        length = tf.shape(x)[1]
        x = self.embedding(x)
        # This factor sets the relative scale of the embedding and positonal_encoding.
        x *= tf.math.sqrt(tf.cast(self.d_model, tf.float32))
        x = x + self.pos_encoding[tf.newaxis, :length, :]
        return x
```

- Next, it uses the ***Intra-attention*** [26]:
    - It is also known as ***self-attention*** [21], with the first implemenation by the modified LSTMs, to see the inter-relationship between tokens (rather than considering both inputs and outputs as the above-mentioned general or **inter attention**):
    - This allows simulatation of the human reading, since we can find important context rather than memorizing everything by LSTM.
- As a basis for all their attention mechanisms, they defined their own attension mechanisms:
    1. **Scaled dot-product attention**:
\begin{equation}
\text{Attention}\left ( Q, K, V \right ) = \text{softmax}  \left ( \frac{QK^\text{T}}{\sqrt{d_k}} \right )V
\end{equation}
        - Here,
            - $Q$, $K$, and $V$ are respectively query, key, value vectors packed into seperate matrices.
            - $d_k$ is the dimension of the query or key space.
        - Thus, its difference with dot-product attention is:
            - the ***scaling*** by the *scale factor* $1 / \sqrt{d_k}$ that counteract the issue of vanishing gradient of the softmax function (figure below);
<img src="img/scaled.png" width="100px">
<center>Scaled dot-product attention [21].</center>
            - the optional mask that is used for decoder self-attention to avoid leftward flow of information and therefore keeping its auto-regressive property by setting all the values of the illegal connectin to −∞  (e.g., the self.attention with masking as shown below).
<img src="img/causal2.png" width="200px">
<center>Simplified representation of auto-regressive self-attention (image from tensorflow.org/responsible_ai).</center>
    2. **Multi-head attention**:
\begin{equation}
\text{MultiHead}(Q, K, V ) = \text{Concat} \left ( \text{head}_1, ..., \text{head}_\text{h} \right )W^O\text{,}\quad \text{where}\quad \text{head}_\text{i} = \text{Attention}\left ( QW_{i}^{Q}, KW_{i}^{K}, VW_{i}^{V} \right )
\end{equation}
        - Here,
            - $Q$, $K$, $V$, and $O$ are respectively the superscripts query, key, value, and output;
            - $W$ is the learnable linear projection of each input' or output' vector to the attention model.
        - Thus, each head gives a diverse representation depending on the projection, which by concatenation of with h heads, it can jointly attend to information from different prespectives to then average them (figure below).
<img src="img/multihead.png" width="160px">
<center>Multi-head attention [21].</center>
        - Keras supports `layers.MultiHeadAttention`.
- Now, all the attention mechanism can be defined by using a base attention class, that also has methods for layer noramlization (by `layers.LayerNormalization()`) and resedual connections (by `layers.Add()` instead of `+` that cannot propogated the Keras masks).

```python
class BaseAttention(tf.keras.layers.Layer):
    def __init__(self, **kwargs):
        super().__init__()
        self.mha = tf.keras.layers.MultiHeadAttention(**kwargs)
        self.layernorm = tf.keras.layers.LayerNormalization()
        self.add = tf.keras.layers.Add()


# the cross attention (i.e., between encoder and decoder)

class CrossAttention(BaseAttention):
    def call(self, x, context):
        attn_output, attn_scores = self.mha(
            query=x,
            key=context,
            value=context,
            return_attention_scores=True)
    
        # Cache the attention scores for plotting later.
        self.last_attn_scores = attn_scores
    
        x = self.add([x, attn_output])
        x = self.layernorm(x)
        
        return x


# The global self attention layer (i.e., without masking)

class GlobalSelfAttention(BaseAttention):
    def call(self, x):
        attn_output = self.mha(
            query=x,
            value=x,
            key=x)
        x = self.add([x, attn_output])
        x = self.layernorm(x)
        return x


# The causal self attention layer (i.e., with masking)

class CausalSelfAttention(BaseAttention):
    def call(self, x):
        attn_output = self.mha(
            query=x,
            value=x,
            key=x,
            use_causal_mask = True)
        x = self.add([x, attn_output])
        x = self.layernorm(x)
        return x
```

```python
# other components:

# the pointwise ffnn.

class FeedForward(tf.keras.layers.Layer):
    def __init__(self, d_model, dff, dropout_rate=0.1):
        super().__init__()
        self.seq = tf.keras.Sequential([
          tf.keras.layers.Dense(dff, activation='relu'),
          tf.keras.layers.Dense(d_model),
          tf.keras.layers.Dropout(dropout_rate)
        ])
        self.add = tf.keras.layers.Add()
        self.layer_norm = tf.keras.layers.LayerNormalization()
  
    def call(self, x):
        x = self.add([x, self.seq(x)])
        x = self.layer_norm(x) 
        return x

# 1 encoder layer

class EncoderLayer(tf.keras.layers.Layer):
    def __init__(self,*, d_model, num_heads, dff, dropout_rate=0.1):
        super().__init__()
    
        self.self_attention = GlobalSelfAttention(
            num_heads=num_heads,
            key_dim=d_model,
            dropout=dropout_rate)
    
        self.ffn = FeedForward(d_model, dff)
    
    def call(self, x):
        x = self.self_attention(x)
        x = self.ffn(x)
        return x


# 1 decoder layer

class DecoderLayer(tf.keras.layers.Layer):
    def __init__(self,
                 *,
                 d_model,
                 num_heads,
                 dff,
                 dropout_rate=0.1):
        super(DecoderLayer, self).__init__()
    
        self.causal_self_attention = CausalSelfAttention(
            num_heads=num_heads,
            key_dim=d_model,
            dropout=dropout_rate)
    
        self.cross_attention = CrossAttention(
            num_heads=num_heads,
            key_dim=d_model,
            dropout=dropout_rate)
    
        self.ffn = FeedForward(d_model, dff)
    
    def call(self, x, context):
        x = self.causal_self_attention(x=x)
        x = self.cross_attention(x=x, context=context)
    
        # Cache the last attention scores for plotting later
        self.last_attn_scores = self.cross_attention.last_attn_scores
    
        x = self.ffn(x)  # Shape `(batch_size, seq_len, d_model)`.
        return x
```

```python
class Encoder(tf.keras.layers.Layer):
    def __init__(self, *, num_layers, d_model, num_heads,
                 dff, vocab_size, dropout_rate=0.1):
        super().__init__()
    
        self.d_model = d_model
        self.num_layers = num_layers
    
        self.pos_embedding = PositionalEmbedding(
            vocab_size=vocab_size, d_model=d_model)
    
        self.enc_layers = [
            EncoderLayer(d_model=d_model,
                         num_heads=num_heads,
                         dff=dff,
                         dropout_rate=dropout_rate)
            for _ in range(num_layers)]
        self.dropout = tf.keras.layers.Dropout(dropout_rate)
  
    def call(self, x):
        # `x` is token-IDs shape: (batch, seq_len)
        x = self.pos_embedding(x)  # Shape `(batch_size, seq_len, d_model)`.
    
        # Add dropout.
        x = self.dropout(x)
    
        for i in range(self.num_layers):
            x = self.enc_layers[i](x)
    
        return x  # Shape `(batch_size, seq_len, d_model)`.


class Decoder(tf.keras.layers.Layer):
    def __init__(self, *, num_layers, d_model, num_heads, dff, vocab_size,
                 dropout_rate=0.1):
        super(Decoder, self).__init__()
    
        self.d_model = d_model
        self.num_layers = num_layers
    
        self.pos_embedding = PositionalEmbedding(vocab_size=vocab_size,
                                                 d_model=d_model)
        self.dropout = tf.keras.layers.Dropout(dropout_rate)
        self.dec_layers = [
            DecoderLayer(d_model=d_model, num_heads=num_heads,
                         dff=dff, dropout_rate=dropout_rate)
            for _ in range(num_layers)]
    
        self.last_attn_scores = None
    
    def call(self, x, context):
        # `x` is token-IDs shape (batch, target_seq_len)
        x = self.pos_embedding(x)  # (batch_size, target_seq_len, d_model)
    
        x = self.dropout(x)
    
        for i in range(self.num_layers):
            x  = self.dec_layers[i](x, context)
    
        self.last_attn_scores = self.dec_layers[-1].last_attn_scores
    
        # The shape of x is (batch_size, target_seq_len, d_model).
        return x
```

- Now, the transformer custom model can be defined with a special attenstion on the possible effect of `_keras_mask` on the loss function, which should be deleted.

```python
class Transformer(tf.keras.Model):
    def __init__(self, *, num_layers, d_model, num_heads, dff,
                 input_vocab_size, target_vocab_size, dropout_rate=0.1):
        super().__init__()
        self.encoder = Encoder(num_layers=num_layers, d_model=d_model,
                               num_heads=num_heads, dff=dff,
                               vocab_size=input_vocab_size,
                               dropout_rate=dropout_rate)
    
        self.decoder = Decoder(num_layers=num_layers, d_model=d_model,
                               num_heads=num_heads, dff=dff,
                               vocab_size=target_vocab_size,
                               dropout_rate=dropout_rate)
    
        self.final_layer = tf.keras.layers.Dense(target_vocab_size)
    
    def call(self, inputs):
        # To use a Keras model with `.fit` you must pass all your inputs in the
        # first argument.
        context, x  = inputs
    
        context = self.encoder(context)  # (batch_size, context_len, d_model)
    
        x = self.decoder(x, context)  # (batch_size, target_len, d_model)
    
        # Final linear layer output.
        logits = self.final_layer(x)  # (batch_size, target_len, target_vocab_size)
    
        try:
            # Drop the keras mask, so it doesn't scale the losses/metrics.
            # b/250038731
            del logits._keras_mask
        except AttributeError:
            pass
    
        # Return the final output and the attention weights.
        return logits
```

```python
# preprocessing by tf.data

MAX_TOKENS=128 # i.e., one token before the [END] token in each sequence.
def prepare_batch(pt, en):
    pt = tokenizers.pt.tokenize(pt)      # Output is ragged.
    pt = pt[:, :MAX_TOKENS]    # Trim to MAX_TOKENS.
    pt = pt.to_tensor()  # Convert to 0-padded dense Tensor

    en = tokenizers.en.tokenize(en)
    en = en[:, :(MAX_TOKENS+1)]
    en_inputs = en[:, :-1].to_tensor()  # Drop the [END] tokens
    en_labels = en[:, 1:].to_tensor()   # Drop the [START] tokens

    return (pt, en_inputs), en_labels


BUFFER_SIZE = 20000
BATCH_SIZE = 64

def make_batches(ds):
    return (
        ds
        .shuffle(BUFFER_SIZE)
        .batch(BATCH_SIZE)
        .map(prepare_batch, tf.data.AUTOTUNE)
        .prefetch(buffer_size=tf.data.AUTOTUNE))


train_batches = make_batches(train_examples)
val_batches = make_batches(val_examples)
```

```python
# setting the hyperparameters
num_layers = 4
d_model = 128
dff = 512
num_heads = 8
dropout_rate = 0.1

# creating the Keras model
transformer = Transformer(
    num_layers=num_layers,
    d_model=d_model,
    num_heads=num_heads,
    dff=dff,
    input_vocab_size=tokenizers.pt.get_vocab_size().numpy(),
    target_vocab_size=tokenizers.en.get_vocab_size().numpy(),
    dropout_rate=dropout_rate)
```

- The original Transformer uses the following learning-rate scheduler that first linearly increase the the learning rate, and then gradually decrease it:

\begin{equation}
lrate = d_{model}^{-0.5} * \min(step{\_}num^{-0.5}, step{\_}num \cdot warmup{\_}steps^{-1.5})
\end{equation}

```python
class CustomSchedule(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, d_model, warmup_steps=4000):
        super().__init__()
        
        self.d_model = d_model
        self.d_model = tf.cast(self.d_model, tf.float32)
        
        self.warmup_steps = warmup_steps
    
    def __call__(self, step):
        step = tf.cast(step, dtype=tf.float32)
        arg1 = tf.math.rsqrt(step)
        arg2 = step * (self.warmup_steps ** -1.5)
        
        return tf.math.rsqrt(self.d_model) * tf.math.minimum(arg1, arg2)



learning_rate = CustomSchedule(d_model)

# for demonstration
plt.plot(learning_rate(tf.range(40000, dtype=tf.float32)))
plt.ylabel('Learning Rate')
plt.xlabel('Train Step')
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate, beta_1=0.9, beta_2=0.98,
                                     epsilon=1e-9)
```

```python
# adding padding mask in the loss and accuracy.

def masked_loss(label, pred):
    mask = label != 0
    loss_object = tf.keras.losses.SparseCategoricalCrossentropy(
        from_logits=True, reduction='none')
    loss = loss_object(label, pred)
    
    mask = tf.cast(mask, dtype=loss.dtype)
    loss *= mask
    
    loss = tf.reduce_sum(loss)/tf.reduce_sum(mask)
    return loss


def masked_accuracy(label, pred):
    pred = tf.argmax(pred, axis=2)
    label = tf.cast(label, pred.dtype)
    match = label == pred
    
    mask = label != 0
    
    match = match & mask
    
    match = tf.cast(match, dtype=tf.float32)
    mask = tf.cast(mask, dtype=tf.float32)
    return tf.reduce_sum(match)/tf.reduce_sum(mask)
```

```python
# training

transformer.compile(
    loss=masked_loss,
    optimizer=optimizer,
    metrics=[masked_accuracy])

transformer.fit(train_batches,
                epochs=20,
                validation_data=val_batches)
```

- two points about inference:
  - we may use `tf.Module` that is in the higher level of Keras models or layers to denfine inference object (not only for a more organized coding but also it has a memory and can be saved), and
  - use `training=False` in the Keras model.

```python
# run inference

class Translator(tf.Module):
    def __init__(self, tokenizers, transformer):
        self.tokenizers = tokenizers
        self.transformer = transformer
    
    def __call__(self, sentence, max_length=MAX_TOKENS):
        # The input sentence is Portuguese, hence adding the `[START]` and `[END]` tokens.
        assert isinstance(sentence, tf.Tensor)
        if len(sentence.shape) == 0:
            sentence = sentence[tf.newaxis]
        
        sentence = self.tokenizers.pt.tokenize(sentence).to_tensor()
        
        encoder_input = sentence
        
        # As the output language is English, initialize the output with the
        # English `[START]` token.

        # The following has only IDs of [START] and [END]
        start_end = self.tokenizers.en.tokenize([''])[0]
        start = start_end[0][tf.newaxis]
        end = start_end[1][tf.newaxis]
        
        # `tf.TensorArray` is required here (instead of a Python list), so that 
        # the dynamic-loop can be traced by `tf.function`.
        output_array = tf.TensorArray(dtype=tf.int64, size=0, dynamic_size=True)
        output_array = output_array.write(0, start)
        
        # input to the loop should be tf.range(max_length) not range(max_length)
        # to keep this loop dynamic.
        for i in tf.range(max_length):
            output = tf.transpose(output_array.stack())
            predictions = self.transformer(
                [encoder_input, output], training=False
                )
            
            # Select the last token from the `seq_len` dimension.
            # Below shape `(batch_size, 1, vocab_size)`.
            predictions = predictions[:, -1:, :]
            
            predicted_id = tf.argmax(predictions, axis=-1)
            
            # Concatenate the `predicted_id` to the output which is given to the
            # decoder as its input.
            output_array = output_array.write(i+1, predicted_id[0])
            
            if predicted_id == end:
                break
        
        output = tf.transpose(output_array.stack())
        # The output shape is `(1, tokens)`.
        text = tokenizers.en.detokenize(output)[0]  # Shape: `()`.
        
        tokens = tokenizers.en.lookup(output)[0]
        
        # `tf.function` prevents us from using the attention_weights that were
        # calculated on the last iteration of the loop.
        # So, recalculate them outside the loop.
        self.transformer([encoder_input, output[:,:-1]], training=False)
        attention_weights = self.transformer.decoder.last_attn_scores
        
        return text, tokens, attention_weights


translator = Translator(tokenizers, transformer)
```

```python
# e.g., trying a translation

def print_translation(sentence, tokens, ground_truth):
    print(f'{"Input:":15s}: {sentence}')
    print(f'{"Prediction":15s}: {tokens.numpy().decode("utf-8")}')
    print(f'{"Ground truth":15s}: {ground_truth}')


sentence = 'este é o primeiro livro que eu fiz.'
ground_truth = "this is the first book i've ever done."

translated_text, translated_tokens, attention_weights = translator(
    tf.constant(sentence))
print_translation(sentence, translated_text, ground_truth)
```

```python
def plot_attention_head(in_tokens, translated_tokens, attention):
    # The model didn't generate `<START>` in the output. Skip it.
    translated_tokens = translated_tokens[1:]
    
    ax = plt.gca()
    ax.matshow(attention)
    ax.set_xticks(range(len(in_tokens)))
    ax.set_yticks(range(len(translated_tokens)))
    
    labels = [label.decode('utf-8') for label in in_tokens.numpy()]
    ax.set_xticklabels(labels, rotation=90)
    
    labels = [label.decode('utf-8') for label in translated_tokens.numpy()]
    ax.set_yticklabels(labels)


head = 0
# Shape: `(batch=1, num_heads, seq_len_q, seq_len_k)`.
attention_heads = tf.squeeze(attention_weights, 0)
attention = attention_heads[head]

in_tokens = tf.convert_to_tensor([sentence])
in_tokens = tokenizers.pt.tokenize(in_tokens).to_tensor()
in_tokens = tokenizers.pt.lookup(in_tokens)[0]

plot_attention_head(in_tokens, translated_tokens, attention)
```

```python
def plot_attention_weights(sentence, translated_tokens, attention_heads):
    in_tokens = tf.convert_to_tensor([sentence])
    in_tokens = tokenizers.pt.tokenize(in_tokens).to_tensor()
    in_tokens = tokenizers.pt.lookup(in_tokens)[0]
    
    fig = plt.figure(figsize=(16, 8))
    
    for h, head in enumerate(attention_heads):
        ax = fig.add_subplot(2, 4, h+1)
        
        plot_attention_head(in_tokens, translated_tokens, head)
        
        ax.set_xlabel(f'Head {h+1}')
    
    plt.tight_layout()
    plt.show()


plot_attention_weights(sentence, translated_tokens, attention_weights[0])
```
