# Support vector machines
## Theory
**SVM** is useful especially for classification of complex small- or medium-sized datasets, regression, and outlier detection. **In linear SVM**, classes are ***linearly separable***, and in **large margin** classifications, the ***decision boundaries*** should be as wide as possible. Here, *off-the-street* instances will not affect SVM and the model is fully determined or supported by the instances on the edges, i.e., ***support vectors***. In **hard margin** classification no instances are located on the street, which only applicable in *linearly separable data* and is very sensitive to the outliners. With **soft margin** classification, limited *margin violations* are allowed.
### Decision Function
For SVM, we assume feature weights vector $\mathbf{w}$ excludes bias $b$.

\begin{equation}
\hat{y}=\ \left\{\begin{matrix}0&if\ \ {\ \mathbf{w}}^T\mathbf{\theta}+b<0\ \ \rightarrow\ \ t^{\left(i\right)}=-1\ \\1&if\ \ {\ \mathbf{w}}^T\mathbf{\theta}+b\geq0\ \ \rightarrow\ \ \ \ t^{\left(i\right)}=0\ \ \\\end{matrix}\right.
\end{equation}

<img src="img/c2.png" width="450px">
<center>With n features, decision function is an n-dimensional hyperplane.</center>

### Quadratic Programming
The linear SVM classifier objective is

\begin{equation}
minimize\left(\mathbf{w},b,\zeta\right)\rightarrow\frac{1}{2}\mathbf{w}^\mathit{T}\mathbf{w}+C\sum_{i=1}^{m} \zeta^{\left ( i \right )}
\end{equation}

\begin{equation}
subject\ to\rightarrow t^{\left(i\right)}\left(\mathbf{w}^\mathit{T}x^{\left(i\right)}+b\right)\geq1-\zeta^{\left(i\right)} \  and \  \zeta^{\left(i\right)}\geq0
\end{equation}

where $\mathbf{w}^\mathit{T}\mathbf{w}$ increases the margin and the *slack variable* $\zeta$ imposes soft margin, and its balance by hard margin SDV is controlled by $C$, which increases the overfitting or decreases the margin. This is a **Quadratic Programming (QP)** problem as it is a convex and quadratic optimization with linear constraints, which can be solved by an off-the-shelf QP solver.
### Duality
The dual problem from the primary optimization problem of a linear model can be used to combined different constraints as it has the necessary conditions. Therefore, As SVM has the required conditions, the **generalized Lagrangian** can be applied to encompass the inequality constraints. For example, for a **hard margin** model, it becomes:

\begin{equation}
\frac{1}{2}\mathbf{w}^\mathit{T}\mathbf{w}-\sum_{i=1}^{m}{\alpha^{\left(i\right)}\left(t^{\left(i\right)}\left(\mathbf{w}^\mathit{T}x^{\left(i\right)}+b\right)-\mathbf{1}\right)}
\end{equation}

Where $\alpha^{\left(i\right)}\geq0$ are **Karush–Kuhn–Tucker (KKT) multipliers**. Under some conditions, it is possible to find the stationary points that are solutions of this optimization problem by its derivatives:

\begin{equation}
minimize\left(\alpha\right)\rightarrow\frac{1}{2}\sum_{i=1}^{m}\sum_{j=1}^{m}{\alpha^{\left(i\right)}\alpha^{\left(j\right)}t^{\left(i\right)}t^{\left(j\right)}{\mathbf{x}^{\left(i\right)}}^T\mathbf{x}^{\left(j\right)}}-\sum_{i=1}^{m}\alpha^{\left(i\right)}
\end{equation}

\begin{equation}
subject\ to\rightarrow\alpha^{\left(i\right)}\geq0
\end{equation}

This can be solved by a QP solver for the vector $\hat{\mathbf{\alpha}}$. Then:

\begin{equation}
\mathbf{w}=\sum_{i=1}^{m}\left({\hat{\alpha}}^{\left(i\right)}t^{\left(i\right)}\mathbf{x}^{\left(i\right)}\right)
\end{equation}

\begin{equation}
\hat{b}=\frac{1}{n_s}\sum_{i=1}^{m}\left(t^{\left(i\right)}-{\hat{\mathbf{w}}}^\mathit{T}\mathbf{x}^{\left(i\right)}\right)
\end{equation}

$n_s$ is used to average bias with respect to all the support vectors. Dual form is faster than primal one when the number of training instances is lower than features and is necessary for kernel tricks.

### Kernelized SVM

According to the ***Mercer’s Theorem**, for any $K\left(\mathbf{a},\mathbf{b}\right)$, which is continuous, symmetric, etc., a mapping function $\emptyset$ exists, where:

\begin{equation}
K\left(\mathbf{a},\mathbf{b}\right)={\emptyset\left(\mathbf{a}\right)}^T\emptyset\left(\mathbf{b}\right)
\end{equation}

For example, in the second-degree polynomial mapping:

<img src="img/CodeCogsEq.jpg" width="350px">

A **kernel** is a function that computes the dot product without transformation $\emptyset$. Common kernels tricks are:

\begin{equation}
\left \{ \begin{matrix}
\mathbf{a}^T\mathbf{b} & \text{Linear}\\ 
\left({\gamma\mathbf{a}}^T\mathbf{b}+r\right)^d & \text{Polynomial}\\ 
exp\left(-\gamma\left \| \mathbf{a} - \mathbf{b}  \right \|^{2}\right) & \text{Gaussian RBF}\\ 
tanh\left({\gamma\mathbf{a}}^T\mathbf{b}+r\right) & \text{Sigmoid}
\end{matrix} \right.
\end{equation}

Here, Gaussian RBF kernel maps to an infinite-dimensional space and also the sigmoid kernel do not respect all the Mercer’s conditions but still works well. This way, the non-linear model can be implemented efficiently so that the prediction is

\begin{equation}
h_{{\hat{\mathbf{w}}}^\mathit{T},\ \ \hat{b}}\left(\emptyset\left(\mathbf{x}^{\left(n\right)}\right)\right)=\mathbf{w}^\mathit{T}\emptyset\left(\mathbf{x}^{\left(n\right)}\right)+\hat{b}=\sum_{i=1}^{m}{\left({\hat{\alpha}}^{\left(i\right)}t^{\left(i\right)}K\left(\mathbf{x}^{\left(i\right)},\mathbf{x}^{\left(n\right)}\right)\right)+\hat{b}}
\end{equation}

Therefore, $\mathbf{w}$ does not need to be solved and it needs the dot product of each new instance $\mathbf{x}^{\left(n\right)}$, for which $\hat{\alpha}>0$, with the support vectors, not all the instances. Similarly, $\hat{b}$ and more complex models can be solved.

### Online SVM
For online ML, use the **hinge loss** function *max(0,\ 1-h)*, to define the following cost function, which can be implemented by GD via any subderivative at t=1:

\begin{equation}
J\left(\mathbf{w},b\right)=\frac{1}{2}\mathbf{w}^\mathit{T}\mathbf{w}+C\sum_{i=1}^{m}\left(max\left(0,\ 1-t^{\left(i\right)}\left(\mathbf{w}^\mathit{T}x^{\left(i\right)}+b\right)\right)\right)
\end{equation}

More efficiently, the kernelized versions could be implemented (e.g. [1], [2]), but generally for large-scale nonlinear problems, neural networks are preferred.

## Implementation
### Linear SVM Classification
The `SVC(kernel="linear", C=float("inf"))` can be used for Large margin classification. The decision boundaries can be then implemented by `.support_vectors_`  , which is not supported `LinearSVC`. Generally, the `C` hyperparameter can regulate the fitting and use `SGDClassifier(, alpha=1/(m*C))` or `LinearSVC` with `loss="hinge"` as general model. Always, use the latter with `dual="False"` when the features are more than training instances.

```python
# Python ≥3.5 is required
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "svm"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
from sklearn.svm import SVC
from sklearn import datasets

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = iris["target"]

setosa_or_versicolor = (y == 0) | (y == 1)
X = X[setosa_or_versicolor]
y = y[setosa_or_versicolor]
```

```python
# SVM Classifier model
svm_clf = SVC(kernel="linear", C=float("inf"))
svm_clf.fit(X, y)
```

```python
def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    w = svm_clf.coef_[0]
    b = svm_clf.intercept_[0]

    # At the decision boundary, w0*x0 + w1*x1 + b = 0
    # => x1 = -w0/w1 * x0 - b/w1
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0]/w[1] * x0 - b/w[1]

    margin = 1/w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin

    svs = svm_clf.support_vectors_
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.plot(x0, decision_boundary, "k-", linewidth=2)
    plt.plot(x0, gutter_up, "k--", linewidth=2)
    plt.plot(x0, gutter_down, "k--", linewidth=2)
```

```python
plot_svc_decision_boundary(svm_clf, 0, 5.5)
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo")
plt.xlabel("Petal length", fontsize=14)
plt.axis([0, 5.5, 0, 2])

save_fig("large_margin_classification_plot")
plt.show()
```

Akin to `SGDClassifier`, `LinearSVC` computational complexity is $O(m\times n)$ and is implemented by **liblinear** library [3] and controlled by `tol` hyperparameter. The SVC class, on the other hand, is implemented by **libsvm** library [4] with the complexity from $O(m^2\times n)$ to $O(m^3\times n)$. Therefore, it is suitable for complex small or medium-sized training sets, especially with ***sparse features***, i.e. each instance is affected by few features. In the latter case, it scales approximately with the average number of nonzero features per instance.

```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(np.float64)  # Iris virginica
```

```python
scaler = StandardScaler()
svm_clf1 = LinearSVC(C=1, loss="hinge", random_state=42)
svm_clf2 = LinearSVC(C=100, loss="hinge", random_state=42)

scaled_svm_clf1 = Pipeline([
        ("scaler", scaler),
        ("linear_svc", svm_clf1),
    ])
scaled_svm_clf2 = Pipeline([
        ("scaler", scaler),
        ("linear_svc", svm_clf2),
    ])

scaled_svm_clf1.fit(X, y)
scaled_svm_clf2.fit(X, y)
```

```python
# Convert to unscaled parameters
b1 = svm_clf1.decision_function([-scaler.mean_ / scaler.scale_])
b2 = svm_clf2.decision_function([-scaler.mean_ / scaler.scale_])
w1 = svm_clf1.coef_[0] / scaler.scale_
w2 = svm_clf2.coef_[0] / scaler.scale_
svm_clf1.intercept_ = np.array([b1])
svm_clf2.intercept_ = np.array([b2])
svm_clf1.coef_ = np.array([w1])
svm_clf2.coef_ = np.array([w2])

# Find support vectors (LinearSVC does not do this automatically)
t = y * 2 - 1
support_vectors_idx1 = (t * (X.dot(w1) + b1) < 1).ravel()
support_vectors_idx2 = (t * (X.dot(w2) + b2) < 1).ravel()
svm_clf1.support_vectors_ = X[support_vectors_idx1]
svm_clf2.support_vectors_ = X[support_vectors_idx2]
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(10,2.7), sharey=True)

plt.sca(axes[0])
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^", label="Iris virginica")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs", label="Iris versicolor")
plot_svc_decision_boundary(svm_clf1, 4, 5.9)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.title("$C = {}$".format(svm_clf1.C), fontsize=16)
plt.axis([4, 5.9, 0.8, 2.8])

plt.sca(axes[1])
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
plot_svc_decision_boundary(svm_clf2, 4, 5.99)
plt.xlabel("Petal length", fontsize=14)
plt.title("$C = {}$".format(svm_clf2.C), fontsize=16)
plt.axis([4, 5.9, 0.8, 2.8])

save_fig("regularization_plot")
```

### Nonlinear SVM Classification
`PolynomialFeatures()` can increase another feature to make them possibly separable. E.g. Moon dataset.

<img src="img/v1.png" width="450px">
<center>Using PolynomialFeatures() for nonlinear SVC.</center>

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.15, random_state=42)
```

```python
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

polynomial_svm_clf = Pipeline([
        ("poly_features", PolynomialFeatures(degree=3)),
        ("scaler", StandardScaler()),
        ("svm_clf", LinearSVC(C=10, loss="hinge", random_state=42))
    ])

polynomial_svm_clf.fit(X, y)
```

```python
def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)

plot_predictions(polynomial_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])

save_fig("moons_polynomial_svc_plot")
plt.show()
```

More efficiently, `SVC(kernel="poly")` can work without finding the polynomial form and consequently is not prone to combinatorial explosion of the number of features. Here, `coef0` hyperparameter increases the effect of the high-degree polynomials, which can be regularized along the other parameters, firstly by coarse then finer grid searches.

```python
from sklearn.svm import SVC

poly_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))
    ])
poly_kernel_svm_clf.fit(X, y)
```

```python
poly100_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="poly", degree=10, coef0=100, C=5))
    ])
poly100_kernel_svm_clf.fit(X, y)
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(10.5, 4), sharey=True)

plt.sca(axes[0])
plot_predictions(poly_kernel_svm_clf, [-1.5, 2.45, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.4, -1, 1.5])
plt.title(r"$d=3, r=1, C=5$", fontsize=18)

plt.sca(axes[1])
plot_predictions(poly100_kernel_svm_clf, [-1.5, 2.45, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.4, -1, 1.5])
plt.title(r"$d=10, r=100, C=5$", fontsize=18)
plt.ylabel("")

save_fig("moons_kernelized_polynomial_svc_plot")
plt.show()
```

Features can also be added by a ***similarity function*** with respect to a ***landmark***. E.g., with **Gaussian Radial Basis Function (RBF)**, the similarity features can be illustrated as:

<img src="img/v2.png" width="450px">
<center>Using RBF for nonlinear SVC.</center>

Commonly, landmarks are located on each m instance, and then the dataset will have **m** features, which should not be high. However the kernel trick can be used by `SVC(kernel="rbf")`. The hyperparameters `gamma` is typically the inverse of the SDV of the RBF. Use this kernel when the training set is not too large.

```python
rbf_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
    ])
rbf_kernel_svm_clf.fit(X, y)
```

```python
from sklearn.svm import SVC

gamma1, gamma2 = 0.1, 5
C1, C2 = 0.001, 1000
hyperparams = (gamma1, C1), (gamma1, C2), (gamma2, C1), (gamma2, C2)

svm_clfs = []
for gamma, C in hyperparams:
    rbf_kernel_svm_clf = Pipeline([
            ("scaler", StandardScaler()),
            ("svm_clf", SVC(kernel="rbf", gamma=gamma, C=C))
        ])
    rbf_kernel_svm_clf.fit(X, y)
    svm_clfs.append(rbf_kernel_svm_clf)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10.5, 7), sharex=True, sharey=True)

for i, svm_clf in enumerate(svm_clfs):
    plt.sca(axes[i // 2, i % 2])
    plot_predictions(svm_clf, [-1.5, 2.45, -1, 1.5])
    plot_dataset(X, y, [-1.5, 2.45, -1, 1.5])
    gamma, C = hyperparams[i]
    plt.title(r"$\gamma = {}, C = {}$".format(gamma, C), fontsize=16)
    if i in (0, 1):
        plt.xlabel("")
    if i in (1, 3):
        plt.ylabel("")

save_fig("moons_rbf_svc_plot")
plt.show()
```

### Regression
Here, the objective is reversed: it puts the instances on street while reducing margin violations (i.e., *off* the street), which is controlled by `epsilon` hyperparameter. As the model is not affected by adding more instances within the margin, it is called ***ϵ-insensitive***. Similarly, both `LinearSVR` and `SVR` can be used.

<img src="img/v3.png" width="450px">
<center>Effect of epsilon.</center>

```python
np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)
y = (4 + 3 * X + np.random.randn(m, 1)).ravel()
```

```python
from sklearn.svm import LinearSVR

svm_reg = LinearSVR(epsilon=1.5, random_state=42)
svm_reg.fit(X, y)
```

```python
np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + np.random.randn(m, 1)/10).ravel()
```

```python
from sklearn.svm import SVR

svm_poly_reg = SVR(kernel="poly", degree=2, C=100, epsilon=0.1, gamma="scale")
svm_poly_reg.fit(X, y)
```
