# Introduction to Machine learning

## Basics

### Definitions

**Machine Learning (ML)**: A computer program learning from *experience E* by *task T* if its *performance P* improves. —T. M., 1997

E.g. Spam filter:
- T → flag spam.
- E → training data. a *training set* is formed by *training instances* (or *samples*).
- P → ratio of correctly classified emails. This particular measure is called *accuracy*, often used in classification.

ML is great when:

- problems require a lot of *fine-tuning* or complex algorithms,
- programs are adaptable to the new data,
- new insights about complex and large data are needed.

ML types:

- By **learning supervision**:
    1. **Supervised**: ML is fed also with solutions (***labels***), e.g.
        + **Classification**: E.g. ***Class*** of emails (spam or ham)
        + **Regression**: predicts a ***target*** numeric value by ***features (predictors)***. A feature may mean ***attribute*** (data type) or an attribute plus its value (***label***), e.g. mileage = 1500.
        + Prediction of values may imply a regression model, e.g.:
            - **Univariate**: single-value prediction (vs. ***multivariate***)
            - **Multiple regression**: prediction with multiple features
        + **Regression for classification**: e.g. *Logistic Regression*
    2. **Unsupervised**: unlabeled training (without teacher), e.g.
        + **Visualization**
        + **Dimensionality reduction**: E.g. using correlations to merge features into one, i.e., ***feature extraction***, mostly used in preprocessing ML.
        + **Clustering**: E.g. detection of groups of similar visitors. In in ***hierarchical clustering***, each group is subdivided into smaller groups.
        + **Anomaly detection**: e.g., ***outlier detection*** (and removing) before learning, e.g., for unusual credit card transactions, and manufacturing defects.
        + **Novelty detection**: detection of new outliers with a ***clean*** training data, which is not polluted by outliers.
        + **Association rule learning**: e.g., it may show some products are bought together in a supermarket.
    3. **Semisupervised**: As labeling is time-consuming and costly, few labels are available, e.g. people labeling in Google Photos.
    4. **Reinforcement**: A learning system (***agent***), observing the environment, performing actions to get ***rewards/penalties*** (or *negative rewards*). Then, the best strategy (***policy***) is learnt to get the most reward, e.g., robot walk.

- By **Incremental learning**: 
    1. **Batch Learning**: without incremental learning typically ***offline*** (not on the *live* system)
    2. **Online (incremental) learning**: with learning on fly and sequentially by either individually or in small groups called ***mini batches***. In such a ***live*** system, the bad data should be monitored, e.g., by an anomaly detection algorithm. For ***huge data*** (with rapidly changing or continuous flow of data), batch learning working on servers (by **MapReduce** techniques [1]) or online learning may be used.
    3. **Out-of-core learning** that can deal with huge datasets by loading part of data on other repositories, is usually done *offline* or *online*.
    
- By **ML generalization**: 
    1. **Instance-based**: ML first learns by hearth some examples, then using a ***measure of similarity*** generalize it.
    2. **Model-based**: ML makes predictions by models (e.g., by a linear regression model). Data should not be *noisy* (partly random) and should have a *trend*. Model selection can be handled by a ***performance*** measure:
        - **Utility function** (or **fitness function**): how good it is. 
        - **Cost function**: how good it is, for e.g., it is used in linear regression as distance between model’s predictions and the training example and the ***objective*** is to minimize it.
    3. In ML, models may also refer to a fully trained system, etc. Overall, models are simplified versions of the observations by assumptions. According to *No Free Lunch (NFL) theorem* [2], no *priori* guaranteed model exists excelling the others.
    
***Learning rate*** is the inverse of ***ML inertia*** showing the sensitivity to the noise or sequences of nonrepresentative data points (outliers).

### Main ML challenges

Quality of data is more important than datasets, but a proper ML algorithm can work more economically [3].
- Bad data:
    - **Nonrepresentative data**, for e.g., Outliners may affect the generalization of the ML system. Too small samples may have ***sampling noise*** (i.e., accidental outliners), and too large samples can be nonrepresentative with sampling bias. E.g., nonresponse bias in polls.
    - **Poor-quality data and measurements** may cause errors, outliers, and noise. Fixing them, for e.g., by removing outliners and fixing missing data, is usually time-consuming.
    - **Irrelevant features** are discarded by feature engineering: 
        - **Feature selection**: selection of the useful features
        - **Feature extraction**
        - Creating new features by gathering new data
- Bad algorithm (over and underfitting): Overgeneralizing data can be fixed by
    - Simplifying model of either the algorithm or attributes,
    - gathering more data,
    - reducing the noise, and
    - constraining the model (**regularization**) by tuning hyperparameters (parameters of a learning algorithm, not of the model). E.g. by tweaking the *degrees of freedom* (DOF) of regression model. (e.g. between 1 and 2 DOF)

### Performance Measure

If $m$: instances, $y$: label, $\hat{\mathbf{y}}$: (“y-hat”) prediction, $h$: ***hypothesis*** → ${\hat{\mathbf{y}}}^{\left(i\right)}=h\left(\mathbf{x}^{\left(i\right)}\right)$, and

\begin{equation}
\mathbf{x}^{\left(\mathbf{i}\right)}=\left(\begin{matrix}\mathbf{first}\ \mathbf{attribute}\\\vdots\\\mathbf{last}\ \mathbf{attribute}\\\end{matrix}\right)
\end{equation}

$\mathbf{L}^\mathbf{P}$ ($l_p$, $p$-norm) of an error vector $\mathbf{v}=\left|\mathbf{y}-\hat{\mathbf{y}}\right|$:

\begin{equation}
l_p=\left(\sum_{i=1}^{m}\left(\mathbf{v}^{\left(i\right)}\right)^p\right)^{1/p}
\end{equation}

Table below listed the norms, where $\mathbf{X}$ is the matrix of attributes:

\begin{equation}
\mathbf{X}=\left(\begin{matrix}\left(\mathbf{x}^{\left(\mathbf{1}\right)}\right)^\mathbf{T}\\\vdots\\\left(\mathbf{x}^{\left(\mathbf{m}\right)}\right)^\mathbf{T}\\\end{matrix}\right)
\end{equation}


| $\mathbf{L}^\mathbf{P}$ or $\left \|  \right \|_p$ | statistical equivalence | description | AKA |
| --- | --- | --- | --- |
| $\mathit{l}_\mathit{0}$ | non-zero counting norm | not a norm! | sparsity |
| $\mathit{l}_\mathbf{1}$ or $\mathbf{MAE}\left(\mathbf{X},\mathit{h}\right)$ | average absolute deviation | not unique | (for regression cost function) Mean Absolute Error, Quadratic loss, Manhattan Distance or Taxicab norm |
| $\mathit{l}_\mathbf{2}$ ,$\left \|  \right \|_p$ , $\mathbf{MSE}$, or $\mathbf{RMSE}\left(\mathbf{X},\mathit{h}\right)$ | standard deviation | unique and sensitive to outliners |  (for regression cost function) Root Mean Square Error, *Euclidean norm*, Mean Squared Error |
| $\mathit{l}_\infty$ | maximum deviation | unique, absolute value of the maximum component | l-infinity, max norm |

### Testing and validating

Statistically, **data snooping** bias appears when a trend in the data is found by chance after deep search. **Generalization errors** (or **out-of-sample error**) are extracted by *test sets*, but **training (in-sample) errors** are derived by *training sets*. Overfitting leads to high Generalization errors but low training errors.

While typically 20% of data were *held out* for tests, larger dataset may *hold out* less.

**Validation set** (AKA **development set** or **dev set**) is a ***held-out set*** from the *full training set* used for evaluation of the performance of the several candidate models (with various hyperparameters) trained on the reduced training set. After this **holdout validation**, the best model is trained on the *full training set*, and then, the generalization error is estimated by the test set. This way, the generalization error is not adapted to that particular test set and may not be affected by new data unless it involves:

- too small validation set, which leads to selection of a suboptimal model by imprecise evaluations, and
- too large validation set, causing small training set, solved by **cross-validation**: each model is evaluated by many small validation sets, and the evaluations are averaged.

Once filtering out (near-)duplicates, **representative data** should be *shuffled*, which is necessary for some ML algorithms, as they perform poorly with similar instances in a row, but it is sometimes troublesome, e.g. time series data, such as stock market prices or weather conditions. Then, the training set is split into the validation and test sets. Then, **train-dev set** is held out from the main training set while maintaining the distribution to evaluate the problems of ***high variance (overfitting)***, ***high bias (underfitting)***, and ***data mismatch*** (owing to *poor preprocessing* and more *complex* training set).

## Scikit-Learn Design

The API of Scikit-Learn is designed by following principles:
- ***Consistency***: All objects share consistent interfaces:
    - ***Estimator***: Anything that estimates on dfs (e.g. imputer), generated by a constructor. `fit()` method performs such estimations, which are controlled by hyperparameters.
    - ***Transformers***: Estimators that also `transform()` a df, which all also supports (sometimes more optimized method), `fit_transform()`.
    - ***Predictors***: Any estimators that can `predict()`; e.g. `LinearRegression`. while `score()` measures the prediction quality, some of the predictors have a method for measuring the confidence of their predictions.
- ***Inspection***: Estimators’ hyperparameters and their learned values are accessible without and with an underscore (e.g., `imputer.strategy` and `imputer.statistics_`)
- ***Nonproliferation of classes***: No new data types are created.
- ***Composition***: building blocks can be combined *Pipeline*.
- ***Sensible defaults***: it has reasonable default parameters.

### Create a test set

For illustration only, as Sklearn has `train_test_split()`:

Manually, the test set can be obtained by a random function as `np.random.permutation(len(data))` unless the dataset is not updated.

```python
# to make this notebook's output identical at every run
np.random.seed(42)
```

```python
def split_train_test(data, test_ratio):
    
    #np.random.shuffle is not applicable in Pandas
    
    shuffled_indices = np.random.permutation(len(data))
    
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
len(train_set), len(test_set)
```

For an updating dataset, we may use ***cyclic redundancy check***, by which the changes in the raw blocks of memory can be checked. Then, if the instances have unique IDs, *crc32*  can be used, producing hash values for each ID and comparing it with a ratio of the maximum hash value. (Unique IDs can be assigned by `df.reset_index()`, or  by making a new column that should be deleted)

```python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
housing_with_id = housing.reset_index()   # adds an `index` column
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")

test_set.head()
```

For **custom transformation** classes (e.g. for attributes cleanup or combination) that can work in sklearn pipelines use the base class `TransformerMixin`. Use the base class `BaseEstimator` to get hyperparameter tuning methods (`get_params()` and `set_params()`), and avoid `*args` or `**kargs`. Here, `fit()` should return `self`. 

```python
from sklearn.base import BaseEstimator, TransformerMixin

# column index
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

Always **scale features** in each column of *training* df, as with all the transformations (table below), typically by a `Pipeline` constructor, which takes a list of name/estimator pairs. All but the last of them should be transformers (i.e., the `fit_transform()` method).

| Feature scaling method | Definition | Specification | Sk-learn code  |
| --- | --- | --- | --- |
| **Min-max scaling** (as **normalization**) | $\frac{x-\min(x)}{\max(x)-\min(x)}$ | Simpler and better | `MinMaxScaler`<br> `feature_range` (to change the default range from 0 to 1) |
| **standardization** | $\frac{x-\bar{x}}{\sigma}$ | Less sensitive to outliners | `StandardScaler` |

```python
# let's build a pipeline to preprocess the numerical attributes:

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
    ])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

Use `ColumnTransformer` to apply the different pipelines on different columns. In this case, `"passthrough"` (string type) can be used instead of a transformer to keep that column untouched.

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])

housing_prepared = full_pipeline.fit_transform(housing)

housing_prepared.shape
```

In old versions, we used `FeatureUnion` with some tricks, which is the parallel implementation of pipelines. Then, the two generated arrays can be compared by `np.allclose`.

```python
from sklearn.base import BaseEstimator, TransformerMixin

# Create a class to select numerical or categorical columns 
class OldDataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
```

```python
num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

# without 'selector' it cannot apply on each column separately.

old_num_pipeline = Pipeline([
        ('selector', OldDataFrameSelector(num_attribs)),
        ('imputer', SimpleImputer(strategy="median")),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
    ])

old_cat_pipeline = Pipeline([
        ('selector', OldDataFrameSelector(cat_attribs)),
        ('cat_encoder', OneHotEncoder(sparse=False)),
    ])
```

```python
from sklearn.pipeline import FeatureUnion

old_full_pipeline = FeatureUnion(transformer_list=[
        ("num_pipeline", old_num_pipeline),
        ("cat_pipeline", old_cat_pipeline),
    ])
```

```python
old_housing_prepared = old_full_pipeline.fit_transform(housing)

np.allclose(housing_prepared, old_housing_prepared)
```

Theses metrics without cross-validation are useless.

 To implement cross-validation, use `cross_val_score`, splitting the df into random **folds**. The result is an array of **scores**, by a utility function, which are the opposite of MSE and, as other series, can be better compared by their *mean* and $\sigma$ by a function or `.describe()`.

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```

```python
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels,
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
# or use:

pd.Series(lin_rmse_scores).describe()
```

`RandomForestRegressor` works on top of *Decision Trees* by averaging out their prediction on random subsets of df. ML model grouping is *ensemble* ML, which may advance and fine-tune ML further.

```python
# n_estimators=100 is the default

from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
forest_reg.fit(housing_prepared, housing_labels)
```

```python
housing_predictions = forest_reg.predict(housing_prepared)
forest_mse = mean_squared_error(housing_labels, housing_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
from sklearn.model_selection import cross_val_score

forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels,
                                scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

Next, instead of fiddling with the ML data, use `GridSearchCV` and `RandomizedSearchCV` to experiment and apply cross-validation, respectively manually or automatically, on possible combinations of hyperparameter values (*space*), features combinations and outliner handling strategies, etc. Most `GridSearchCV` parameters, e.g. `best_params_`, `best_estimator_`, and `cv_results_` are available if a multi-metric scoring is not used. Otherwise, `refit` parameter should be used (it refit an estimator using the best parameters).

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    # try 12 (3×4) combinations of hyperparameters
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    # then try 6 (2×3) combinations with bootstrap set as False
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

forest_reg = RandomForestRegressor(random_state=42)

# train across 5 folds, that's a total of (12+6)*5=90 rounds of training

grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(housing_prepared, housing_labels)
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

`RandomizedSearchCV`, which can use `randint`, computes the `feature_importances_` revealing the uninformative features, which can be compared along with each attribute’s name.

### Launch and maintain the system

Get the solution ready for production (e.g., polish the code, write documentation and tests, and so on), then deploy the sklearn model to production environment by saving the model on a web server. The model may be wrapped within a dedicated web service to optimize deployment (e.g., reducing the upgrading, load balancing and scaling issues) that can query through a REST API — It is an HTTP-based API that follows some conventions, such as the standard HTTP verbs (`GET`, `POST`, `PUT`, and `DELETE`), and also, use JSON — and allowing for application of any languages in the web app. Alternatively, the model can be stored on cloud, such as, Google Cloud AI Platform by uploading the `joblib` file to Google Cloud Storage (GCS), taking JSON requests from the website.

Monitoring code should be written to check your system’s live performance continuously and trigger alerts upon it drops, which can be a steep drop or a gentle decay, possibly because the model tend to ***rot*** over time. E.g. for a recommender system of a shop, the number of purchased recommended products can be monitored. But sometime human analysis in needed, for example, by sending especially the suspicious data to human raters.

A complex maintaining system may (semi-)automatically collect fresh data, and fine-tune the hyperparameters, update and compare the models. It keeps backups of the models permitting rolling back to the previous version if the model gets corrupted. Also, the quality of input data should be checked as it may be missing some of the important features or degrade the system by a poor-quality signal. Finally, several subsets of trainset can be created (e.g., the most recent items and the rest) to better evaluate it on the different versions.
